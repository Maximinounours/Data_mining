```python
# ========= IMPORTS =========

import urllib.request
import json
import pandas as pd
import requests
from PIL import Image
from PIL.ExifTags import TAGS
from bs4 import BeautifulSoup
import random
import IPython.display as display
import matplotlib.pyplot as plot

import os
import shutil

LIMIT_IMG_CAT = 5
```

```python
# ========= Créer les répertoires et fichiers utiles =========
try:
   os.mkdir('./images')
except:
    shutil.rmtree('./images')
    os.mkdir('./images')

open('./data.json', 'w').close()
```

```python
# ========= Differentes categories d'images =========

categorie = ["https://commons.wikimedia.org/wiki/Category:Quality_images_of_Acinonyx_jubatus",
             "https://commons.wikimedia.org/wiki/Category:Quality_images_of_Bison_bison",
             "https://commons.wikimedia.org/wiki/Category:Quality_images_of_Kobus_kob",
             "https://commons.wikimedia.org/wiki/Category:Quality_images_of_cats",
             "https://commons.wikimedia.org/wiki/Category:Quality_images_of_Panthera_uncia",
             "https://commons.wikimedia.org/wiki/Category:Featured_pictures_of_Accipitriformes",
             "https://commons.wikimedia.org/wiki/Category:Featured_pictures_of_Haliaeetus",
             "https://commons.wikimedia.org/wiki/Category:Featured_pictures_of_Anseriformes"
            ]
```

```python

#Notre BDD
data = {}
```

```python
# ========= LES FONCTIONS =========


#Telecharge l'image dont le lien est ids i save == True
#Sauvegarde le chemin relatif vers cette image
def save_image(id, save):
    #save : pour enregistrer l'image ou non (mettre False si l'image est deja telechargee)
    
    #Nom de notre image dans le dossier images
    nom_image = id.split("/")[-1]
    #Recupere la page html de l'image et ses donnees
    req = requests.get(id)
    soup = BeautifulSoup(req.text, 'html.parser')
    #Recupere le lien vers l'image a telecharger
    lien_image = soup.find('div', {'id' : 'file'}).find('img').attrs['src']
    #Permet de sauvegarder une image dans nos donees locales si voulu
    if save:
      f = open("./images/" + nom_image,'wb')
      f.write(requests.get(lien_image).content)
      f.close()
    
    #Ajoute le tag nom_image dans noter base de doneee
    add_data_tag(id, 'nom_image',"./images/" + nom_image)


#Ajoute une entree dans notre base de donnee pour une image
def add_data(id):
    data[id] = {}

#Ajoute pour l'image id une association tag/valeur
#l'id de notre base de donnee est le lien vers la page web de l'image
def add_data_tag(id, tag, valeur):
    if data.get(id) is None:
        data[id] = {tag : valeur}
    else:
        data[id][tag] = valeur
        
#Ecrit la BDD data dans le fichier JSON
def ecriture_json():
    with open('./data.json', 'w') as fichier:
        json.dump(data, fichier)

#Retire une entree de la BDD (pas du fichier JSON)
def remove_data(id):
    data.pop(id)
    #ecriture_json()

def get_orientation(width, height):
    format_image = "paysage"
    if width == None or height == None:
        return None
    if(width < height):
        format_image = "portrait"
    elif width == height:
        format_image = "carre"
    return format_image
        
```

```python
# ========= Beautiful Soup, recuperation des images =========


#Faire une boucle for dans les differentes categories
for i in categorie:
    lien = i
    req = requests.get(lien)
    soup = BeautifulSoup(req.text, 'html.parser')

    #Recup tous les liens des images (id dans notre BDD)
    MAX_IMG = len(soup.find('div', {'id':'mw-category-media'}).find_all('img'))
    nb_img = 0

    #On parcourt les images du div contenant les contenant toutes
    #Valable uniquement pour les pages "Category:" sur Wikispecies
    for a in soup.find('div', {'id':'mw-category-media'}).find_all('img'):

        #Limite a LIMIT_IMG_CAT images par categorie
        if nb_img == LIMIT_IMG_CAT or nb_img == MAX_IMG:
            break

        #Lien partiel vers l'image
        a = a.find_parent('a').attrs['href']

        #Recup extension de l'image
        b = a.split(".")[-1]

        #On ne garde que .png ou .jpg
        if (b=="png" or b=="jpg"):

            #Compteur pour ne pas en avoir trop
            nb_img += 1

            #Lien complet vers la page de l'image
            lien = 'https://commons.wikimedia.org'+ a

            #Ajoute dans notre base de donnee une entree pour le l'image trouvee
            add_data(lien)

#Ecriture dans le fichier JSON
ecriture_json()
```

```python
#Ajout des tags dans le dictionnaire
for lienImg in data:
    
    reqImg = requests.get(lienImg)
    soup = BeautifulSoup(reqImg.text, 'html.parser')
    summary=soup.find('div', {'id' : 'mw-imagepage-content'})
    image=soup.find('div', {'id' : 'file'}).find('img')
    try:
        date=int(summary.find('time',{'class':'dtstart'}).attrs['datetime'].split('-')[0])
    except: 
        date=None
    
    try:
        height=int(image.attrs['data-file-height'])
    except: 
        height=None
        
    try:
        width=int(image.attrs['data-file-width'])
    except: 
        width=None
        
    try:
        extension=soup.find("span", class_="mime-type").get_text().split('/')[1]
    except: 
        extension=None
        
    tags=[]
    for category in soup.findAll('div', {'id' : 'mw-normal-catlinks'})[0].findChildren("a")[1:]:
        tags.append(category.attrs['title'].split(':')[1]) 
        
    add_data_tag(lienImg, 'date', date)
    add_data_tag(lienImg, 'width', width)
    add_data_tag(lienImg, 'orientation', get_orientation(width,height))
    add_data_tag(lienImg, 'height', height)
    add_data_tag(lienImg, 'extension', extension)
    add_data_tag(lienImg, 'categorie', tags[0])
#Ecriture dans le fichier JSON
ecriture_json()
```

```python
# ========= Sauvegarde les images =========

for id in data:
    save_image(id, True)
    
ecriture_json()
```

```python
# ========= Recuperer les couleurs dominantes de notre image =========

import numpy
from sklearn.cluster import MiniBatchKMeans
import math
import time

def get_exif_field (exif,field) :
  for (clef,valeur) in exif.items():
     if ExifTags.TAGS.get(clef) == field:
        return valeur

n = 4 #clusters Kmeans

cpt = 0

for id in data:
    #1 a 2s par image

    #start = time.time()
    
    #Erreur potentielle d'ouverture de l'image
    try:
        IMG = Image.open(data[id]['nom_image'])
    except:
        add_data_tag(id, 'color1', None)
        add_data_tag(id, 'color2', None)
        continue

    #Lecture des données EXIF
    img_exif = IMG.getexif()
    
    #Check s'il y a des donnees exif
    if img_exif is None:
        add_data_tag(id, 'color1', None)
        add_data_tag(id, 'color2', None)
        continue

    #Check s'il y a 3 canaux de couleur (image en couleur)
    if len(IMG.getbands()) != 1:

        #Couleurs predominantes avec Kmeans
        numarray = numpy.array(IMG.getdata(), numpy.uint8)

        color = []
        cluster_count = 5 

        clusters = MiniBatchKMeans(n_clusters = cluster_count)
        clusters.fit(numarray)

        npbins = numpy.arange(0, cluster_count + 1)
        histogram = numpy.histogram(clusters.labels_, bins=npbins)
        labels = numpy.unique(clusters.labels_)

        for nbr in range(cluster_count):
            color.append('#%02x%02x%02x' % (
            math.ceil(clusters.cluster_centers_[nbr][0]), 
                math.ceil(clusters.cluster_centers_[nbr][1]),
          math.ceil(clusters.cluster_centers_[nbr][2])))
        
        #Recup les couleurs dominantes
        color1 = color[0]
        color2 = color[1]
    
    #Si image en N&B
    else:
        color1 = '#ffffff'
        color2 = '#000000'

    add_data_tag(id, 'color1', color1)
    add_data_tag(id, 'color2', color2)
    
    #end = time.time()
    #print(end - start)
    
ecriture_json()
```

```python
# ========= CLASS UTILISATEUR =========

class Utilisateur:
    def __init__(self):
        self.id = input("Entrez l'id de l'utilisateur : ")
        self.liked = []
        self.disliked = []
        self.fav_tags = {}
    
    def showStats(self):
        print("ID : " + self.id)
        print("Liked : " + str(self.liked))
        print("Disliked : " + str(self.disliked))
```
